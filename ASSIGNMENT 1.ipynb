```python
# 3. Calculate the mean of each column of the dataset.

print("The mean of column 'SL' is :")
a=df['SL'].mean()
print(round(a,2))
print("The mean of column 'SW' is :")
b=df['SW'].mean()
print(round(b,2))
print("The mean of column 'PL' is :")
c=df['PL'].mean()
print(round(c,2))
print("The mean of column 'PW' is :")
d=df['PW'].mean()
print(round(d,2))
```
